### Diwali Sales Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
Data.head(5)
```

```python
Data.info()
```

```python
Data.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
Data
```

```python
Data.dropna(inplace=True)
Data
```

```python
pd.isnull(Data).sum()
```

```python
Data.info()
```

```python
Data['Amount']=Data['Amount'].astype('int')
```

```python
Data.rename(columns={'Marital_Status':'Shaadi'}, inplace=True)
```

## Exploratory Analysis

## Gender

```python
Data.columns
```

```python
Gender_count=sns.countplot(x= 'Gender',data=Data)
plt.ylabel=''
for bars in Gender_count.containers:
    Gender_count.bar_label(bars)
```

```python
Group_by_Values=Data.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
```

```python
Gender_amount=sns.barplot(x='Gender', y='Amount', data=Group_by_Values)
for i in Gender_amount.containers:
    Gender_amount.bar_label(i)
```

## Age

```python
Age_group=sns.countplot(x='Age Group', data=Data, hue='Gender')
for bars in Age_group.containers:
    Age_group.bar_label(bars)
```

```python
Age_amount=Data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
Age_group_amount=sns.barplot(x='Age Group', y='Amount', data=Age_amount)
```

## State

```python
State=Data.groupby(['State'],as_index=False)['Orders'].sum().sort_values (by='Orders', ascending=False).head(10) 
sns.set({'figure.figsize':(20,5)})
sns.barplot(x='State', y="Orders", data=State)
```

```python
State_amount=Data.groupby(['State'],as_index=False)['Amount'].sum().sort_values (by='Amount', ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
sns.barplot(x='State', y='Amount', data=State_amount)
```
